```python
from transformers import pipeline

# Fungsi untuk mengidentifikasi topik dari teks tertentu
def identifikasi_topik(teks, label_kandidat):
    classifier = pipeline("zero-shot-classification")
    hasil = classifier(teks, candidate_labels=label_kandidat)

    # Tampilkan hasil dalam format yang rapi
    for label, skor in zip(hasil["labels"], hasil["scores"]):
        print(f"Topik: {label} - Tingkat Kepastian: {skor:.2%}")

# Teks yang akan diklasifikasi dan label topik yang mungkin
konten_teks = "Perubahan iklim yang drastis menyebabkan cuaca ekstrem di berbagai belahan dunia."
label_kandidat = ["perubahan iklim", "kesehatan", "ekonomi"]

# Jalankan fungsi identifikasi topik
identifikasi_topik(konten_teks, label_kandidat)
```

```python
from transformers import pipeline

# Fungsi kreatif untuk mengidentifikasi topik berdasarkan teks dan label kandidat
def analisa_topik(teks, label_topik):
    classifier = pipeline("zero-shot-classification")
    hasil = classifier(teks, candidate_labels=label_topik)

    # Menampilkan hasil dengan format yang rapi dan menarik
    print("🔍 Analisis Topik Berdasarkan Teks:")
    print(f"Teks: \"{teks}\"")
    for label, skor in zip(hasil["labels"], hasil["scores"]):
        print(f"Topik: {label} - Kemungkinan: {skor:.2%}")

# Teks yang akan dianalisis dan label kandidat untuk klasifikasi
konten = "Pasar cryptocurrency mengalami volatilitas tinggi setelah berita terbaru mengenai regulasi pemerintah."
label_kandidat = ["keuangan", "cryptocurrency", "regulasi pemerintah"]

# Panggil fungsi analisa_topik
analisa_topik(konten, label_kandidat)
```

```python
from transformers import pipeline

# Fungsi untuk klasifikasi topik dari teks tertentu
def klasifikasi_topik(teks, label_topik):
    classifier = pipeline("zero-shot-classification")
    hasil = classifier(teks, candidate_labels=label_topik)

    # Menampilkan hasil dengan format yang rapi
    print("🧠 Hasil Klasifikasi Topik Kesehatan Mental 🧠")
    print(f"Teks: \"{teks}\"")
    for label, skor in zip(hasil["labels"], hasil["scores"]):
        print(f"Topik: {label} - Tingkat Kepastian: {skor:.2%}")

# Teks baru yang akan dianalisis dan label kandidat
konten = "Pelatihan ini berfokus pada teknik pengelolaan stres dan kesehatan mental di tempat kerja."
label_kandidat = ["kesehatan mental", "manajemen stres", "psikologi"]

# Memanggil fungsi klasifikasi_topik
klasifikasi_topik(konten, label_kandidat)
```

```python
from transformers import pipeline

# Fungsi untuk menghasilkan teks kreatif berdasarkan prompt
def buat_teks(prompt, max_length=50, num_return_sequences=1):
    generator = pipeline("text-generation")
    hasil = generator(prompt, max_length=max_length, num_return_sequences=num_return_sequences)

    # Menampilkan hasil dengan format yang menarik
    print("🌍 Teks yang Dihasilkan 🌍")
    for i, teks in enumerate(hasil, 1):
        print(f"Pilihan {i}:\n{teks['generated_text']}\n")

# Prompt
prompt_teks = "fix this computer will"

# Memanggil fungsi buat_teks
buat_teks(prompt_teks, max_length=60, num_return_sequences=3)
```

```python
from transformers import pipeline

# Inisiasi pipeline untuk Question Answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Konteks
context = "Hydrogen is a clean alternative to methane, also known as natural gas. It's the most abundant chemical element, estimated to contribute 75% of the mass of the universe. Here on earth, vast numbers of hydrogen atoms are contained in water, plants, animals and, of course, humans."
question = "what is Hydrogen?"

# Mendapatkan jawaban dari model
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

# Inisiasi pipeline untuk Question Answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Konteks baru yang lebih menarik dan informatif
context = (
    "Perubahan iklim adalah fenomena global yang ditandai dengan perubahan suhu dan pola cuaca yang "
    "signifikan. Aktivitas manusia, terutama pembakaran bahan bakar fosil, telah meningkatkan konsentrasi "
    "gas rumah kaca di atmosfer, yang mengakibatkan pemanasan global. Konsekuensi dari perubahan iklim "
    "termasuk peningkatan permukaan laut, cuaca ekstrem, dan dampak negatif terhadap ekosistem dan "
    "kehidupan manusia. Upaya untuk mengatasi perubahan iklim mencakup pengurangan emisi, penggunaan "
    "energi terbarukan, dan perlindungan terhadap keanekaragaman hayati."
)
question = "Apa yang dimaksud dengan perubahan iklim?"

# Mendapatkan jawaban dari model
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

# Inisiasi pipeline untuk analisis sentimen
classifier = pipeline("sentiment-analysis")

# Prompt
result = classifier("kamu baik ; dia membantu orang!")

# Menampilkan hasil analisis sentimen
print(f"Sentimen: {result[0]['label']}, Skor: {result[0]['score']:.2f}")
```

```python
from transformers import pipeline

# Inisiasi pipeline untuk ringkasan teks
summarizer = pipeline("summarization")

# Teks
text = """
Perkembangan teknologi dan pendidikan teknik di Amerika Serikat telah mengalami perubahan drastis dalam beberapa tahun terakhir.
Jumlah lulusan dalam disiplin teknik tradisional seperti teknik mesin, sipil, listrik, kimia, dan aeronautika mengalami penurunan.
Universitas-universitas terkemuka di Amerika kini lebih fokus pada studi ilmu teknik dan sains, sehingga mengurangi tawaran
mata kuliah teknik yang berkaitan dengan infrastruktur dan lingkungan. Sementara itu, negara-negara dengan ekonomi berkembang cepat
seperti China dan India terus mendorong pengajaran teknik, menghasilkan lulusan teknik tradisional jauh lebih banyak dibandingkan Amerika.
Negara-negara industri lainnya pun menjaga tingkat kelulusan mereka, sementara Amerika mengalami penurunan serius dalam jumlah
lulusan teknik yang berkualitas.
"""

# Mendapatkan ringkasan dari teks
summary = summarizer(text, max_length=200, min_length=30, do_sample=False)

# Menampilkan hasil ringkasan
print("Ringkasan:")
print(summary[0]['summary_text'])
```

```python
from transformers import pipeline

# Inisiasi pipeline untuk penerjemahan
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")

# Teks
text_to_translate = "Kreativitas adalah bagian penting dari hidupku; aku sangat suka menggambar dan mengekspresikan diri melalui seni."

# Mendapatkan hasil terjemahan
result = translator(text_to_translate)

# Menampilkan hasil terjemahan
print("Hasil Terjemahan:")
print(result[0]['translation_text'])
```
